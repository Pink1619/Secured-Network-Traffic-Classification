```python
import os
import glob as glob
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
folder_vpn = "data\\pcap_to_csv\\vpn"
filepaths_vpn = [i for i in glob.glob(folder_vpn + "\*")]
print(filepaths_vpn)
```

```python
label = 'vpn'
 
for path in filepaths_vpn:
    filename = (path.split('.')[0]).split('\\')[-1]
    print(filename)
    dummy_df = pd.read_csv(path)
    dummy_df['Label'] = label
    dummy_df.to_csv(os.path.join("data\\all_csvs\\" + filename + '.csv'), index=False)
```

```python
folder_non_vpn = "data\\pcap_to_csv\\non_vpn"
filepaths_non_vpn = [i for i in glob.glob(folder_non_vpn + "\*")]
print(filepaths_non_vpn)
```

```python
label = 'non_vpn'
 
for path in filepaths_non_vpn:
    filename = (path.split('.')[0]).split('\\')[-1]
    print(filename)
    dummy_df = pd.read_csv(path)
    dummy_df['Label'] = label
    dummy_df.to_csv(os.path.join("data\\all_csvs\\" + filename + '.csv'), index=False)
```

```python
Allfilepaths = [i for i in glob.glob("data\\all_csvs" + "\*")]
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in Allfilepaths])
# suffle 
combined_csv=combined_csv.sample(frac=1)
#export to csv
combined_csv.to_csv( "data\\vpn-non_vpn.csv", index=False, encoding='utf-8-sig')
```
